```python
#Comparaci√≥n de peso, g√©nero a trav√©s de los a√±os. 
women_athletes = athletes[athletes.Sex == 'F']
men_athletes = athletes[athletes.Sex == 'M']

fig, ax = plt.subplots()
women_athletes_weight = women_athletes.pivot_table(index=['Year'], values=['Weight'],aggfunc=np.mean)
men_athletes_weight = men_athletes.pivot_table(index=['Year'], values=['Weight'],aggfunc=np.median)
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.lineplot(x=women_athletes_weight.index, y=women_athletes_weight.Weight, marker='o')
sns.lineplot(x=men_athletes_weight.index, y=men_athletes_weight.Weight,marker='o')
plt.title('Peso promedio por a√±o',fontsize=22)
plt.xticks(men_athletes_weight.index,rotation=90)
plt.ylabel('Kg')
plt.xlabel('A√±o')
plt.show()


```

```python
#Comparaci√≥n de altura, g√©nero a trav√©s de los a√±os. 
women_athletes = athletes[athletes.Sex == 'F']
men_athletes = athletes[athletes.Sex == 'M']

fig, ax = plt.subplots()
women_athletes_height = women_athletes.pivot_table(index=['Year'], values=['Height'],aggfunc=np.mean)
men_athletes_height = men_athletes.pivot_table(index=['Year'], values=['Height'],aggfunc=np.median)
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.lineplot(x=women_athletes_height.index, y=women_athletes_height.Height, marker='o',color='slateblue')
sns.lineplot(x=men_athletes_height.index, y=men_athletes_height.Height,marker='o',color='mediumseagreen')
plt.title('Altura promedio por a√±o',fontsize=22)
plt.xticks(men_athletes_height.index,rotation=90)
plt.ylabel('Cm')
plt.xlabel('A√±o')
plt.show()
```

```python
#Edades
edades = athletes.pivot_table(index=['Age'], values=['ID'],aggfunc={'ID':'count'})
edades_20 = edades[edades.index <= 20].agg(['sum'])
edades_21a29 = edades[(edades.index > 20) & (edades.index < 30)].agg(['sum'])
edades_30a39 = edades[(edades.index > 29) & (edades.index < 40)].agg(['sum'])
edades_40a49 = edades[(edades.index > 39) & (edades.index < 50)].agg(['sum'])
edades_50a59 = edades[(edades.index > 49) & (edades.index < 60)].agg(['sum'])
edades_60 = edades[edades.index > 60].agg(['sum'])
```

```python
#Dona de edades
fig, ax = plt.subplots(figsize=(12, 9), subplot_kw=dict(aspect="equal"))

labels = ['<20','21-29','30-39','40-49','50-59','60<']

data = [edades_20['ID'], edades_21a29['ID'], edades_30a39['ID'], edades_40a49['ID'], edades_50a59['ID'], edades_60['ID']]

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.6", fc="w", ec="k", lw=0.42)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0)


for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(labels[i], xy=(x, y), xytext=(.75*np.sign(x), 1.2*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Edades", fontsize=22)
plt.show()
```

```python
#Medallas
tipos_medallas = list(set(athletes.Medal))
tipos_medallas

medallas = athletes.groupby(["Team", "Year"])["Medal"].count().unstack()
medallas.head()

paises_medallas = athletes.groupby(["Team"])["Medal"].count().sort_values(ascending=False).head(15)
lista_paises= [pais for pais in paises_medallas.index]
lista_paises

med = medallas.loc[lista_paises]
med.loc["Resto del mundo"] = medallas.loc[medallas.index.difference(lista_paises)].sum()
med = med[::-1]
```
